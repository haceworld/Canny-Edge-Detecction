```python
#Canny Edge Detection 

import cv2
import numpy as np
import matplotlib.pyplot as plt
#img=cv2.imread(r"C:\Users\Aabha\Downloads\sylwia-bartyzel-eU4pipU_8HA-unsplash (1).jpg")
img=cv2.imread("65_1.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

canny=cv2.Canny(img,50, 60)#threshold values
#canny=cv2.Canny(img,30,40)#threshold values
titles=['image','canny']
images=[img,canny]
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
filename = "CECECED_M2"  + ".jpeg"
cv2.imwrite(filename, canny)
```
